### Load Web Page with Current Registration List

```python
import requests
from bs4 import BeautifulSoup
import time

#load the results page with all events
html_page = requests.get('https://www.ibjjfdb.com/ChampionshipResults/1221/PublicRegistrations?lang=en-US')

#check is there is an issue with the html page
try:
    html_page.raise_for_status()
except Exception as exc:
    print('There was a problem: %s' % (exc))

# script to see how long it takes to parse the page
start = time.time()
soup_main = BeautifulSoup(html_page.content, 'lxml') #adding .text here removes markup
end = time.time()
print(end - start) #the time it takes to scrape the page in lxml (less than html)
```

```python
#download all tournaments with divisions and fighter results
import pandas as pd
data = []

for page in events:
    response = requests.get(page)
    soup = BeautifulSoup(response.content, 'lxml')

    event = soup.find('h2').get_text(strip=True)
    for i in soup.find_all('div', {'class': 'col-xs-12'}):
        for s in i.find_all(['h3','h4','div'],recursive=False):
            if s.name == 'h3':
                typ = re.sub('\s+', ' ', s.get_text(strip=True))
            elif s.name == 'h4':
                cat = re.sub('\s+', ' ', s.get_text(strip=True))
            elif s.name == 'div':
                divs = s.find_all('div')
                if len(divs) > 0:
                    for di in divs:
                        info = re.sub('\s+', ' ', di.get_text(strip=True))
                else:
                    info = re.sub('\s+', ' ', s.get_text(strip=True))
                data.append((event,typ,cat,info))

df = pd.DataFrame(pd.DataFrame(data, columns=['Tournament','Results','Divisions','Fighter']))
df
```
